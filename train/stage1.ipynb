```python
import os
import zipfile

zip_folder = "/kaggle/input/Depi_data"
extract_to = "/kaggle/working/"

os.makedirs(extract_to, exist_ok=True)

zip_files = [f for f in os.listdir(zip_folder) if f.endswith(".zip")]

for zip_file in zip_files:
    zip_path = os.path.join(zip_folder, zip_file)
    zip_extract_folder = os.path.join(extract_to, os.path.splitext(zip_file)[0])
    os.makedirs(zip_extract_folder, exist_ok=True)
    
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(zip_extract_folder)
```
