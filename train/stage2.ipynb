```python
import os
import zipfile

output_dir = "/kaggle/working/dataset"
paths = {
        "train_images" : f"{output_dir}/train/images", 
        "train_labels" : f"{output_dir}/train/labels",
        "val_images" : f"{output_dir}/val/images",
        "val_labels" : f"{output_dir}/val/labels",
    } 
for p in paths.values(): os.makedirs(p, exist_ok=True)
```

```python
train_images_zip_files = ["dawn and dusk.zip", "day.zip" , "night.zip"]
train_labels_zip_files = ["dawn and dusk labels.zip", "day labels.zip" , "night labels.zip"]
val_images_zip_files = ["val.zip"]
val_labels_zip_files = ["val labels.zip"]

zip_folder = "/kaggle/input/Depi_data"

for zip_file in train_images_zip_files:
    zip_path = os.path.join(zip_folder, zip_file)
    zip_extract_folder = os.path.join("/kaggle/working/dataset/train/images", os.path.splitext(zip_file)[0])
    os.makedirs(zip_extract_folder, exist_ok=True)
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(zip_extract_folder)

for zip_file in train_labels_zip_files:
    zip_path = os.path.join(zip_folder, zip_file)
    zip_extract_folder = os.path.join("/kaggle/working/dataset/train/labels", os.path.splitext(zip_file)[0])
    os.makedirs(zip_extract_folder, exist_ok=True)
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(zip_extract_folder)

for zip_file in val_images_zip_files:
    zip_path = os.path.join(zip_folder, zip_file)
    zip_extract_folder = os.path.join("/kaggle/working/dataset/val/images", os.path.splitext(zip_file)[0])
    os.makedirs(zip_extract_folder, exist_ok=True)
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(zip_extract_folder)

for zip_file in val_labels_zip_files:
    zip_path = os.path.join(zip_folder, zip_file)
    zip_extract_folder = os.path.join("/kaggle/working/dataset/val/labels", os.path.splitext(zip_file)[0])
    os.makedirs(zip_extract_folder, exist_ok=True)
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(zip_extract_folder)
```
